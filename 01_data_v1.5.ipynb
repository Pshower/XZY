```python
import pandas as pd
import numpy as np
import akshare as ak
import time
from multiprocessing.dummy import Pool as ThreadPool
from q_tools.data import *
```

```python
# 下载数据
start_time = time.time()
down_index()
code = download_index_codes()
with ThreadPool(8) as pool:
    pool.map_async(down_data, code) #加载线程
    pool.close()  #
    pool.join()
end_time=time.time()
print('程序运行时间：{:.2f}秒'.format(end_time - start_time))
```

```python
# 合并整理数据
start_time = time.time()
code = download_index_codes()
df=clean_data(code[0])
path_file = '01_data/fill/%s.csv' % (code[0])
df.to_csv(path_file)
seq=[code[0]]*df.shape[0]
df.insert(0, 'code',seq)
for i in code[1:]:
    df1=clean_data(i)
    path_file = '01_data/fill/%s.csv' % (i)
    df1.to_csv(path_file)
    seq=[i]*df1.shape[0]
    df1.insert(0, 'code',seq)
    df1['code']=i
    df=pd.concat([df,df1])
df.to_csv('01_data/total.csv')
end_time = time.time()
print('程序运行时间：{:.2f}秒'.format(end_time - start_time))
```
